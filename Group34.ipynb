```python
#FINAL CODE

import pandas as pd

import datetime

def GetDate():
    now = datetime.datetime.now()
    print("Current date and time")
    print(now.strftime("%Y-%m-%d %H:%M"))
    return

url = "https://www.teamrankings.com/mlb/trends/ats_trends/"
ats = pd.read_html(url,header=0)[0]
    
url2 = "https://www.teamrankings.com/mlb/trends/ou_trends/"
overund = pd.read_html(url2,header=0)[0]

url3 = "https://www.teamrankings.com/mlb/schedules/"
games = pd.read_html(url3,header=0)[0]
matchups = games['Matchup']

ats["Cover_per"] = ats["Cover %"].str.replace('%', "")

ats['Cover_float'] = ats["Cover_per"].astype('float64')

overund["Over_per"] = overund["Over %"].str.replace('%', "")

overund['Over_float'] = overund["Over_per"].astype('float64')

yats = ats['Team'][ ats['Cover_float'] >= 60.0]

nats = ats['Team'][ ats['Cover_float'] <= 40.0]

yover = overund['Team'][ overund['Over_float'] >= 60.0]

yunder = overund['Team'][ overund['Over_float'] <= 40.0]


GetDate()
print("\n""Today's games are: ", "\n", matchups)

print("\n""You should bet the OVER on: ", "\n\n", yover)

print("\n""You should bet the UNDER on: ", "\n\n", yunder)

print("\n""You should bet the SPREAD on: ", "\n\n", yats)

print("\n""You should bet AGAINST the spread on: ", "\n\n", nats)






```
